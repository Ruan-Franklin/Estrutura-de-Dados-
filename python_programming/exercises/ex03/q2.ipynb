## Imprimir a tabuada do n√∫mero 3 (3 x 1 = 1 - 3 x 10 = 30)


```python
contador = 0
while contador < 10:
    contador+=1
    resultado=3*contador
    print("3 x {} = {}".format(contador, resultado))
```
