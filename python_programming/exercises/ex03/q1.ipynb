```python
# Ex 1: Ler 5 notas e informar a média usando for e while
```

```python
soma = 0
for i in range(5):
    nota = float(input("Digite a nota {}: ".format(i+1)))
    soma += nota

media = soma / 5
print("A média das notas é:", media)

```

```python
elementos = 0
soma = 0
while(elementos <= 4):
    nota = float(input("Digite a nota"))
    soma += nota
    elementos += 1
    print(elementos)

media = soma / 5
print("A média das notas é:", media)
```
