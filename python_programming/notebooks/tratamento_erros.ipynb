```python
while True:
    try:
        p = int(input("Digite um número: "))
    except ValueError:
        print("Valor inválido! Digite um número inteiro.")
    except KeyboardInterrupt:
        print("\nEntrada interrompida pelo usuário.")
        break
    else:
        print(f"Você digitou o número {p}.")
        break
```
