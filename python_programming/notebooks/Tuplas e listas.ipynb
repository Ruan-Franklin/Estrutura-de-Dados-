## Aula sobre tuplas e listas em Python

## Listas

```python
l1 = ["Homo Sapiens", "Canis Familiares", "Felis Catus"]
l2 = ["Xenopus Laeve", "Melanoleuca"]
l2.append("Neymar")
print(l1)
print(l2)
```
